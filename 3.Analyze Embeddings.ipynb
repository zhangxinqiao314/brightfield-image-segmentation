## set jVAE parameters
* en_original_step_size: input image data size
* pool_list: pooling parameters for each maxpool layer
* embedding_size: number of embedding channels
* conv_size: number of filters for each CNN layer
* de_original_step_size: the original image size for the decoder to recover to the input image data size
* up_list: upsample parameters for each upsampling layer

```python
[xdim,ydim] = logdata.shape[-2:]
```

```python
en_original_step_size=[xdim,ydim]
pool_list=[4,4,2]

de_original_step_size = [8,8]
up_list = [2,2,4]
embedding_size=8
conv_size =128

batch_size=16
```

```python
encoder = model_utils.Encoder(original_step_size=en_original_step_size,
                  pool_list=pool_list,
                  embedding_size=embedding_size,
                  conv_size=conv_size,device=device).to(device)

decoder = model_utils.Decoder(original_step_size=de_original_step_size,
                  up_list=up_list,pool_list=pool_list,
                  embedding_size=embedding_size,
                  conv_size=conv_size).to(device)
model = model_utils.VAE(encoder, decoder).to(device)

# optimizer
optimizer = optim.Adam(model.parameters(), lr=3e-5)
```

```python
from scipy.signal import find_peaks
from scipy.ndimage import rotate
from skimage.measure import profile_line
from scipy.signal import savgol_filter
from scipy.fft import fft, ifft,fftfreq
```

```python
fig, ax = utils.layout_fig(3,mod=2,size=6)

## Make mask of filtered
mask1 = resize(mask,emb_obj.orig_filtered[t].shape)
x=np.linspace(0,599,600)
y=900+x
ax[0].imshow(mask1*emb_obj.orig_filtered[t])
ax[0].plot(x,y,'r-')
ax[0].set_title('Masked Filtered BF image')

## Make Profile Line
line = profile_line(mask1*emb_obj.orig_filtered[t],
            [0,900],[599,0])
profile = savgol_filter(line,71,2)
ax[1].plot(line) 
ax[1].plot(profile,linewidth=3)
ax[1].legend(['Line Profile','Filtered'])
ax[1].set_title('Line Profile')
ax[1].set_xlabel('Line Position')
ax[1].set_ylabel('Amplitude (a.u.)')
ax[1].grid()

## Make fft and plot in freq space
fft_profile=fft(1-profile)
cleaned=abs(fft_profile)
cleaned[cleaned>1000]=0

N = len(fft_profile)
# sample spacing
T = 1.0 / 800.0
x = np.linspace(0.0, N*T, N, endpoint=False)
y = 1-profile
yf = fft(y)
xf = fftfreq(N, T)[:N//2]
y_plot = 2/N * np.abs(cleaned[0:N//2])

ax[2].plot(xf, y_plot)
ax[2].grid()
ax[2].set_title('Line Profile FFT')
ax[2].set_xlabel('Period of Needles (pixels/needle)')
ax[2].set_ylabel('Amplitude (a.u.)')
ax[2].scatter(xf[38],y_plot[38],s=40,color='red')
ax[2].annotate(f'(px/needle ~{xf[38]:.2f})',(xf[38]*1.5,y_plot[38]*.95))
```
